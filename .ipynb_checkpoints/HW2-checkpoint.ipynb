1) A little introductory brain teaser. Which is more probable when rolling 2 six-sided dice: rolling snake eyes (two ones) or rolling sevens (dice sum to seven)? What is the ratio of the probabilities?

Rolling snake eyes is a question of rolling a one on the first die AND a one on the second die, so the equation neccessary is 
$P(A+B) = P(A)P(B)$.

```python
P_A = 1/6
P_B = 1/6

P_AB = P_A * P_B
print(P_AB)
```

Rolling sevens is a question of rolling a one AND a six or a two AND a five or a three AND a four and so on. Each of these possibilities is a sum, so this means we need to use the equation of probabilities for sums  $\sum\limits_{z}=P_{A}(z)P_{B}(x-z)$ where z represents the value on a given side of the first die, and x is the sum we want to achieve. Since we already know the probability of each combination, we can multiply that probability by the number of times it could occur.

```python
P_seven = 6*P_AB
print(P_seven)
```

```python
print(P_AB/P_seven)
```

2) Following what we did in class show how to use the convolution operator to determine the probability of the sum of 2 six sided dice. Do both analytically (math & counting) and numerically (computer program). Beware the implicit definition of the values (x-axis on histogram).

```python
import numpy as np
```

```python
b = 1/6
c = np.array([1,2,3,4,5,6,5,4,3,2,1])
P = np.empty(11, dtype=object)
for i in range(11):
    P[i] = c[i]*b**2
print(P)
```

```python
a = np.empty(6, dtype=object)
a[:] = 1/6
v = a.copy()

P_sum = np.convolve(a,v)
print(P_sum)
```

4) Repeat 2, and graph the average of 10 dice. Is this is a Gaussian distribution? Explain in depth.

```python
P_3 = np.convolve(a,P_sum)
P_4 = np.convolve(a,P_3)
P_5 = np.convolve(a,P_4)
P_6 = np.convolve(a,P_5)
P_7 = np.convolve(a,P_6)
P_8 = np.convolve(a,P_7)
P_9 = np.convolve(a,P_8)
P_10 = np.convolve(a,P_9)

import matplotlib
import matplotlib.pyplot as plt
```

```python
# plotting a histogram
fig, ax = plt.subplots(1, 1)
#ax.hist(P_10, 50, density=True)   # d is the distribution defined above, 50 is the # of bins in the histogram, and density=True returns the probability density
plt.tick_params(labelsize = 24)
plt.xlim([0.95,6.05])
x = np.linspace(0.95, 6.05, 51)
ax.plot(x, P_10, linewidth = 8, alpha = 0.7)  # alpha designates opaqueness of curve
plt.show()
```

```python
# plotting a histogram
fig, ax = plt.subplots(1, 1)
plt.yscale('log')
plt.tick_params(labelsize = 24)
plt.xlim([0.95,6.05])
x = np.linspace(0.95, 6.05, 51)
ax.plot(x, P_10, linewidth = 8, alpha = 0.7)  # alpha designates opaqueness of curve
plt.show()
```

Convolving this many times does appear to produce a Gaussian, as we can see from the shape of the logplot above.

5) Show that the sum and average of an initially Gaussian distribution is also a Guassian (can be analytic or numerical). How does the standard deviation of the resulting sum or average Guassian change? This is a hugely important result. Explore what this means for integrating a signal over time.

```python
import scipy
from scipy import stats
# 100k points from Gaussian/Normal distribution
d = stats.norm.rvs(loc = 5., scale = 0.01, size = 100000)
```

```python
fig, ax = plt.subplots(1, 1)
plt.tick_params(labelsize = 24)
plt.xlim([4.95,5.05])
x = np.linspace(4.95, 5.05, 1000)
ax.plot(x, 2*stats.norm.pdf(x, loc = 5., scale = 0.01), linewidth = 8, alpha = 0.7)  # alpha designates opaqueness of curve
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
plt.yscale('log')
plt.tick_params(labelsize = 24)
plt.xlim([4.95, 5.05])
x = np.linspace(4.95, 5.05, 1000)
ax.plot(x, 2*stats.norm.pdf(x, loc = 5., scale = 0.01), linewidth = 8, alpha = 0.7)
plt.show()
```

The above graphs show that adding two Gaussians still makes a Gaussian.

```python
C = np.convolve(x,d)
```

```python
fig, ax = plt.subplots(1, 1)
plt.tick_params(labelsize = 24)
plt.xlim([4.95,5.05])
x = np.linspace(4.95, 5.05, C.size)
ax.plot(x, C, linewidth = 8, alpha = 0.7)  # alpha designates opaqueness of curve
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
plt.tick_params(labelsize = 24)
plt.yscale('log')
plt.xlim([4.95,5.05])
x = np.linspace(4.95, 5.05, C.size)
ax.plot(x, C, linewidth = 8, alpha = 0.7)  # alpha designates opaqueness of curve
plt.show()
```

The plots above show that the average of a Gaussian is not a Gaussian. This shows that the standard deviation has increased.
