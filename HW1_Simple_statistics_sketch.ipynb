```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
import scipy
from scipy import stats

# setting the size of the plot
plt.rcParams["figure.figsize"] = (15,10)
```

```python
# 100k points from Gaussian/Normal distribution
d = stats.norm.rvs(loc = 5., scale = 0.01, size = 100000)
```

```python
# plotting a histogram
fig, ax = plt.subplots(1, 1)
ax.hist(d, 50, density=True)   # d is the distribution defined above, 50 is the # of bins in the histogram, and density=True returns the probability density
plt.tick_params(labelsize = 24)
plt.xlim([4.95,5.05])
x = np.linspace(4.95, 5.05, 1000)
ax.plot(x, stats.norm.pdf(x, loc = 5., scale = 0.01), linewidth = 8, alpha = 0.7)  # alpha designates opaqueness of curve
plt.show()
```

```python
ax = plt.hist(d, 50)
plt.yscale('log')  # sets a logarithmic scale on the y-axis
plt.tick_params(labelsize = 24)
plt.xlim([4.95, 5.05])
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(d, 50, density=True)
plt.yscale('log')
plt.tick_params(labelsize = 24)
plt.xlim([4.95, 5.05])
x = np.linspace(4.95, 5.05, 1000)
ax.plot(x, stats.norm.pdf(x, loc = 5., scale = 0.01), linewidth = 8, alpha = 0.7)
plt.show()
```
